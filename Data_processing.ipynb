```python
# Import dependencies
import os
import pandas as pd
```

```python
# importing housing data
house_df = pd.read_excel('Resources/clean_House_prices.xlsx')
house_df.head()
```

```python
# importing income data
income_df = pd.read_excel('Resources/CO_income.xlsx')
income_df
```

```python
cleaned_house_df = house_df.groupby(['CountyName']).mean()
cleaned_house_df
```

```python
counties_list = list(cleaned_house_df.index.values)
counties_list
```

```python
period_list = income_df.periodyear.unique()
period_list
```

```python
header = ["County", "Year", "House_price"]
import csv
with open("Resources/house_price.csv", "w", newline="") as house_data:
    csvwriter = csv.writer(house_data )
    csvwriter.writerow(header)
    for county in counties_list:
        for period in period_list:
            house_price = cleaned_house_df.loc[county, period]
            csvwriter.writerow([county, period, house_price])
```
