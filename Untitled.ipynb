```python
import os
import pandas as pd
```

```python
# importing data
house_df = pd.read_excel('Resources/clean_House_prices.xlsx')
house_df.head()
```

```python
cleaned_house_df = house_df.groupby(['CountyName']).mean()
cleaned_house_df
```

```python
# Create CountyName and CountyID column
cleaned_house_df.reset_index(inplace=True)
cleaned_house_df.reset_index(inplace=True)
cleaned_house_df = cleaned_house_df.rename(columns={"index": "CountyID"})
```

```python
# Save house prices df in to data_visualize for visualizing purpose
cleaned_house_df.to_csv('Resources/data_visualize.csv', index=False)
cleaned_house_df
```
